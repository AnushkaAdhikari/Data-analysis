#  Data exlporation

#### 
1. Complaint by Borough

```python
import matplotlib.pyplot as plt # for ploting/ visualization
# Create a new figure for the pie chart with specific size
plt.figure(figsize=(5, 6))
# Count the number of complaints for each borough and plot it as a pie chart
df['Borough'].value_counts().plot(kind='pie',autopct='%1.1f%%',startangle=90,colors=['pink', 'lightblue', 'lightgreen', 'orange', 'brown'])
# Set title
plt.title('Complaints Distribution by Borough')
plt.ylabel('')  # Makes the pie chart a perfect circle and clear the label
# Display the chart
plt.show()
```

#### 2. Most common complaint type

```python
plt.figure(figsize=(10,4))#Create a new figure for plot with specific size 

#Count the occurrences of each complaint type, select the top 10 most frequent, and plot as a bar chart
df['Complaint Type'].value_counts().head(10).plot(kind='bar', color='skyblue',ec='blue')

plt.title('Top 10 Complaint Types')#set title of the chart
#set level for x-axis and y-axis
plt.xlabel('Complaint Type')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45)
plt.show()#display the plot
```

#### 3. Monthly complaint trend

```python
df['Month'] = df['Created Date'].dt.month # provide month with the help of date
plt.figure(figsize=(10, 6))#Create a new figure for plot with specific size 
# Plot a bar chart showing the number of complaints for each month
df['Month'].value_counts().sort_index().plot(kind='bar', color='lightgreen',ec='green')

plt.title('Complaints per Month')#set title of the chart
#set level for x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Number of Complaints')
plt.show()#display the plot
```

### 4. Average response time by Borough

```python
# Group the data set by borough and caluculate mean for request_closing_time for each group and sort result in desending order
avg_closing_time = df.groupby('Borough')['Request_Closing_Time'].mean().sort_values(ascending=False)

#Create a new figure for plot with specific size 
plt.figure(figsize=(10, 4))

# Create average closing time as a bar chart with suitable color
avg_closing_time.plot(kind='bar', color='salmon', ec='red')
plt.title('Average Request Closing Time by Borough')#set title of the chart

plt.xlabel('Borough')#set level for x-axis
plt.ylabel('Avg. Closing Time (Hours)')#set level for y-axis

plt.show()#display the plot
```

### 5. arrange complaint

```python
# Group the data by Complaint Type and Borough, and calculate the average Request_Closing_Time for each group
avg_closing_by_complaint_location = df.groupby(['Complaint Type', 'Borough'])['Request_Closing_Time'].mean().unstack()

# Plotting the average closing time using a bar chart
avg_closing_by_complaint_location.plot(kind='bar', figsize=(15, 5), colormap='viridis')

# Set the chart title and axis labels
plt.title('Average Request Closing Time by Complaint Type and Borough')
plt.xlabel('Complaint Type')
plt.ylabel('Avg. Closing Time (Hours)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
#plt.tight_layout()
plt.show()
```

# Test

### Test 1

##### Null Hypothesis (H0): the average response response time is same across all complaint type
##### Alternative Hypothesis (H1): Average response time is different from at least one complaint type

```python
import pandas as pd
from scipy.stats import f_oneway

# Step 1: Convert Request_Closing_Time (timedelta) to hours (float)
df['Request_Closing_Hours'] = df['Request_Closing_Time'].dt.total_seconds() / 3600

# Step 2: Group the numeric column by Complaint Type
grouped_data = df.groupby('Complaint Type')['Request_Closing_Hours'].apply(list)

# Step 3: Perform one-way ANOVA (requires multiple groups of numeric data)
anova_result = f_oneway(*grouped_data)

# Step 4: Print the test results
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

### Test 2

##### H0= Complaint type and location (borough) are independent
##### H1= Complaint type and location (borough) are not independent

```python
from scipy.stats import chi2_contingency

# Create a contingency table of Complaint Type vs Borough
# Each cell in the table will count the number of complaints for that type in that borough
contingency_table = pd.crosstab(df['Complaint Type'], df['Borough'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results

print("p-value:", p)
```
